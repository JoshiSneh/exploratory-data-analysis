```python
import pandas as pd
```

```python
df = pd.read_csv('diwali_sales.csv', encoding='iso-8859-1')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
mean = df["Amount"].mean()
```

```python
mean
```

```python
df["Amount"].fillna(mean, inplace=True)
```

```python
df["Amount"].isnull().sum()
```

```python
df["Amount"] = df["Amount"].astype(int)
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
df[["Amount","Age"]].describe()
```

```python
df["Gender"].value_counts()
```

```python
import seaborn as sns
```

```python
# palette_colors = {"M": "orange", "F": "blue"}
sns.countplot(x="Gender", data=df, hue="Gender")
```

```python
sales_gen = df.groupby(["Gender"], as_index=False)["Amount"].sum()
```

```python
sales_gen
```

```python
sns.barplot(x="Gender",y="Amount",data=sales_gen,hue="Gender")
```

```python
df["Age Group"].value_counts()
```

```python
sns.countplot(x="Age Group", data=df, hue="Gender")
```

```python
sales_age = df.groupby(["Age Group"], as_index=False)["Amount"].sum()
```

```python
sales_age
```

```python
sns.barplot(x="Age Group",y="Amount",data=sales_age,hue="Age Group")
```

```python
df["State"].value_counts()
```

```python
sns.countplot(x="State", data=df,hue="State")
sns.set_theme(rc={'figure.figsize':(40,5)})
```

```python
sales_state = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False).head(10)
```

```python
sales_state
```

```python
sns.barplot(x="State",y="Orders",data=sales_state,hue="State")
sns.set_theme(rc={'figure.figsize':(25,5)})
```

```python
sales_amount = df.groupby(["State"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False).head(10)
```

```python
sales_amount
```

```python
sns.barplot(x="State",y="Amount",data=sales_amount,hue="Amount")
sns.set_theme(rc={'figure.figsize':(25,5)})
```
