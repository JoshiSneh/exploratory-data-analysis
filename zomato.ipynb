```python
df_country = pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
final_df = pd.merge(df, df_country,on='Country Code')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df["Country"].value_counts()
```

```python
country_name = final_df["Country"].value_counts().index
country_name[0]
```

```python
country_value = final_df["Country"].value_counts().values
country_value
```

```python
## Top 3 countries with most number of zomato orders
sns.barplot(x=country_name[:3], y=country_value[:3],hue=country_name[:3])
```

```python
final_df.columns
```

```python
df_rating = final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index(name="Rating Count")
```

```python
df_rating
```

```python
# Observation:
# When rating is between 4.5 to 4.9 then rating is -- excellent
# When rating is between 4.0 to 4.4 then rating is -- very good
# When rating is between 3.5 to 3.9 then rating is -- good
# When rating is between 2.5 to 3.4 then rating is -- average
# When rating is between 1.8 to 2.4 then rating is -- poor
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=df_rating, hue="Rating text")
sns.set_theme(rc={'figure.figsize':(30,10)})
```
